```python
import cv2 
import pytesseract
from pytesseract import Output
```

```python
img = cv2.imread('data/Inside@18x-100.jpg')
```

```python
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(pytesseract.image_to_string(img))
```

```python
data = pytesseract.image_to_data(img,output_type=Output.DICT)
```

```python
data['text']
```

```python
data['conf']
```

```python
n_text = len(data['text'])
n_text
```

```python
img = cv2.imread('data/invoice-sample.jpg')
for i in range(n_text):

    if int(data['conf'][i])>70:    
        x,y,w,h = data['left'][i],data['top'][i],data['width'][i],data['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import re 
```

```python
date_pattern = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
mail_pattern =	'^[a-z0-9](\.?[a-z0-9]){5,}@example\.net$'
```

```python
img = cv2.imread('data/invoice-sample.jpg')
for i in range(n_text):

    if int(data['conf'][i])>70:    

        if re.match(date_pattern,data['text'][i]):

            x,y,w,h = data['left'][i],data['top'][i],data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('data/invoice-sample.jpg')
for i in range(n_text):

    if int(data['conf'][i])>60:    

        if re.match(mail_pattern,data['text'][i]) or re.match(date_pattern,data['text'][i]):

            x,y,w,h = data['left'][i],data['top'][i],data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```
